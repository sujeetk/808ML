## Feature Visualization

Lets take a look at the images of digits

```python
def show_digit_images(data, labels):
    for i in range(data.shape[0]):
        plt.subplot(1, data.shape[0], i+1)
        plt.imshow(data[i].reshape((28,28)))    
        plt.title(labels[i])
        plt.xticks(())
        plt.yticks(())
for d in range(10):        
    X_d = X_train[y_train == d]
    rsubset = np.random.permutation(X_d.shape[0])[:10]    
    plt.figure()
    show_digit_images(X_d[rsubset], d + np.zeros(X_d.shape[0]))
```

Lets project the data to 2D using PCA and plot them to get a sense of how discriminative the feature are

```python
def visualize_features_2D(X, y):  
    
    from sklearn.decomposition import RandomizedPCA
    from itertools import cycle
    from sklearn.lda import LDA

    colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
    markers = ['+', 'o', '^', 'v', '<', '>', 'D', 'h', 's']
    
    # pca
    pca = RandomizedPCA(n_components=2)
    X_pca = pca.fit_transform(X)

    plt.figure()
    for i, c, m in zip(np.unique(y), cycle(colors), cycle(markers)):
        plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1],
                    c=c, marker=m, label=i, alpha=0.5)        
    _ = plt.legend(loc='best')
    plt.title('Projection of features to 2D using Randomized PCA')
    
    # lda
    lda = LDA(n_components=2)
    X_lda = lda.fit(X, y).transform(X)

    plt.figure()
    for i, c, m in zip(np.unique(y), cycle(colors), cycle(markers)):
        plt.scatter(X_lda[y == i, 0], X_lda[y == i, 1],
                    c=c, marker=m, label=i, alpha=0.5)        
    _ = plt.legend(loc='best')
    plt.title('Projection of features to 2D using Linear Discriminant Analysis')
    
%time visualize_features_2D(X_train, y_train)
```

Lets take a look at the first few principal components

```python
from sklearn.decomposition import RandomizedPCA
pca = RandomizedPCA(n_components=10)
%time X_pca = pca.fit_transform(X_train)
show_digit_images(pca.components_[:10], range(10))
```

Lets try to find the number of components to keep after dimensionality reduction using PCA

```python
pca = RandomizedPCA()
%time X_pca = pca.fit_transform(X_train)

x = range(len(pca.components_))
cumSumVar = np.cumsum(pca.explained_variance_ratio_)
plt.plot(x, cumSumVar)
plt.yticks(np.arange(0.0,1.01,0.05))
plt.xticks(np.arange(0,len(x),50))
plt.xlabel('Number of principal components')
plt.ylabel('Percentage of Variance Explained')
```

```python
percentageOfVarianceToKeep = 0.95
n_components_to_keep = np.min(np.nonzero(cumSumVar> percentageOfVarianceToKeep))
print n_components_to_keep
```

## Classification Models

```python
from sklearn import preprocessing
from sklearn import cross_validation
from sklearn import grid_search
from sklearn import metrics
from sklearn.pipeline import Pipeline

numCVFolds = 5
scoringMetric = 'accuracy'
```

```python
data_test = pd.read_csv('Data/test.csv')
X_test = data_test.values
```

```python
data_scaler = preprocessing.StandardScaler()
dim_red_model = RandomizedPCA(n_components = n_components_to_keep)
```

### Support Vector Machines

```python
from sklearn import svm
from sklearn import linear_model
```

Lets first try linear SVM

```python
linear_svm_pipeline = Pipeline([
                                 ('scaler', data_scaler),
                                 ('reduce_dim', dim_red_model),
                                 ('clf', svm.SVC(kernel='linear'))
])
```

```python
linear_svm_params = {
                     'clf__C': np.logspace(-7, 2, 10)
}
gs = grid_search.GridSearchCV(linear_svm_pipeline, linear_svm_params, cv=numCVFolds, scoring=scoringMetric, verbose=1, n_jobs=4)
#%time gs.fit(X_train, y_train)
```

```python
#sorted(gs.grid_scores_, key=lambda x: x.mean_validation_score)
```

Lets build the model with optimal parameters we just found using grid search: C = 0.01

```python
linear_svm_pipeline = Pipeline([
                                 ('scaler', data_scaler),
                                 ('reduce_dim', dim_red_model),
                                 ('clf', svm.SVC(kernel='linear', C=0.01))
])
```

```python
%time linear_svm_pipeline.fit(X_train, y_train)
```

Lets check the accuracy on the training set

```python
linear_svm_pipeline.score(X_train, y_train)
```

Lets check the accuracy on the validation set

```python
linear_svm_pipeline.score(X_validate, y_validate)
```

Now lets generate the results

```python
%time y_test_linear_svm = linear_svm_pipeline.predict(X_test)
```

```python
res_linear_svm = pd.DataFrame( {'ImageId': 1 + np.arange(X_test.shape[0]), 'Label': y_test_linear_svm} )
%time res_linear_svm.to_csv('results/sub_linear_svm.csv', index=False)
```
